# defining Experiment

```python
l_min = 0.8
l_max = 1.05
l_step = 0.0002 #5000 points per order of l

wavelength = 10**-10
direct = 2.0E8
background = 1

STO = xpy.CrystalStructureCheck(path = xpy.scriptpath+"Perovskites/STOh.str")
```

# Substrate

## sim

```python
substrate = xpy.ThickFilmCheck(STO)
sample = xpy.Sample(substrate)

sub001 = xpy.l_scan(sample, STO, l_min = l_min, l_max = l_max, l_step = l_step, direct = direct, background = background, wavelength = wavelength)
```

## plot

```python
fig = plt.figure(figsize=(16,10), dpi=100)

axes = fig.add_axes([0.0, 0.0, 1.0, 1.0]) # left, bottom, width, height (range 0 to 1)

axes.plot([float(x[0])+0.002 for x in BTOSTOSTO5[1]], [float(a[1]) for a in BTOSTOSTO5[1]], color="red", lw=2, ls='*', marker='+', label='data')
axes.plot([x[0] for x in sub001], [float(a[1]/5) for a in sub001], color="green", lw=2, ls='*', marker='+', label='simulation')

axes.set_xlabel('l')
axes.set_ylabel('int[A.U.]')
axes.set_title('title')
axes.set_yscale("log")
axes.legend()
axes.grid(True)
```

# Electrode

## sim

```python
SRO = xpy.CrystalStructureCheck(path = xpy.scriptpath+"Perovskites/SROs.str")
electrode = xpy.ThinFilm(SRO,47)
SRO.lattice_parameters[0] = SRO.lattice_parameters[1] = 3.925*10**-10
SRO.lattice_parameters[2] = 3.999*10**-10
electrode.delta = 0.05
sample = xpy.Sample(substrate, electrode)

e001 = xpy.l_scan(sample, STO, l_min, l_max, l_step, direct = direct, background = background, wavelength = wavelength)
```

## plot

```python
fig = plt.figure(figsize=(16,10), dpi=100)

axes = fig.add_axes([0., 0., 1., 1.]) # left, bottom, width, height (range 0 to 1)

axes.plot([float(x[0])+0.002 for x in BTOSTOSRO2[1]], [float(a[1]) for a in BTOSTOSRO2[1]], color="red", lw=2, ls='*', marker='+', label='data')
axes.plot([x[0] for x in e001], [float(a[1]/19) for a in e001], color="green", lw=2, ls='*', marker='+', label='simulation')

axes.set_xlabel('l')
axes.set_ylabel('int[A.U.]')
axes.set_title('title')
axes.set_yscale("log")
axes.legend()
axes.grid(True)
```

### Animation

```python
path = r'/home/benjaminbein/workespace/git/x-ray_simulator/Perovskites/BTOSTO.str'
xpy.create_structure(path, 3.925, 3.991, 'Ba', 'Ti', 'Sr', 'Ti', 2.3, 6.0, 0.01,0)
BTOSTO = xpy.CrystalStructureCheck(path = xpy.scriptpath + "Perovskites/BTOSTO.str")
superlattice = xpy.Layer(BTOSTO)
superlattice.delta = 0.000
sample = xpy.Sample(substrate, electrode, superlattice)
s001s = [0]
s001s[0] = xpy.l_scan(sample, STO)
    
fig = plt.figure(figsize=(16,10), dpi=100)

ax = fig.add_axes([0.05, 0.05, 0.95, 0.95]) # left, bottom, width, height (range 0 to 1)

ax.set_xlabel('l')
ax.set_ylabel('int[A.U.]')
ax.set_title('BTO/STO Superlattice on SRO electrode')
ax.set_yscale("log")
ax.set_ylim(10**(-2), 10**6)
ax.set_xlim(0.8, 1.05)
ax.legend()
ax.grid(True)

data, = ax.plot([], [], color="red", lw=1, ls='*', marker='+', label = 'data')
sim, = ax.plot([], [], color="green", lw=1, ls='*', marker='+', label = 'sim')
BTOscans = 10
STOscans = 26
scans = BTOscans
grow = "BTO"

test = [grow, scans, 0, ax.text(0.975, 0.02, r"Growing: %s, Bilayers: %i" % (grow , 0), fontsize=20, color="black")]

def init():
    data.set_data([], [])
    sim.set_data([], [])
    
def update(n):
    # n = frame counter
    data.set_data([float(x[0])+0.002 for x in BTOSTOSRO2[n+1]], [y[1] for y in BTOSTOSRO2[n+1]])
    
    b = float((n+1))/36
    bilayers = (n+1)//36
    if ((n+1) == test[1]):
        if (test[0] == "BTO"):
            test[0] = "STO"
            test[1] += STOscans
        else:
            test[0] = "BTO"
            test[1] += BTOscans
    layers = b*(2.3+6.0)
    if (float(int(layers)) == layers):
        layers = int(layers)
    else:
        layers = int(layers)+1
        
    if (layers != test[2]):
        xy=(0.5,0.5)
        test[3].set_text(r"Growing: %s, Bilayers: %i" % ( test[0] , bilayers))
        xpy.create_structure(path, 3.925, 3.991, 'Ba', 'Ti', 'Sr', 'Ti', 2.3, 6.0, b)
        BTOSTO = xpy.CrystalStructureCheck(path = xpy.scriptpath + "Perovskites/BTOSTO.str")
        superlattice = xpy.Layer(BTOSTO)
        sample = xpy.Sample(substrate, electrode, superlattice)
        s001s[0] = xpy.l_scan(sample, STO)
    
    test[2] = layers
    sim.set_data([x[0] for x in s001s[0]], [y[1]/40 for y in s001s[0]])
    
anim = animation.FuncAnimation(fig, update, init_func=init, blit=True, frames=len(BTOSTOSRO2))

anim.save('001s.mp4', fps=1, writer="ffmpeg", codec="libx264")
```
