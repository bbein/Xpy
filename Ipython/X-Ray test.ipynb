```python
def load_sim(path):
    data = []
    f = open(path, 'r')
    line=f.readline()
    line=f.readline()
    while (line):
        words = line.split(' ')
        if words:
            if (len(words) > 1):
                data.append([words[2],words[3]])
        line=f.readline()
    f.close()
    return data

def load_data(path):
    data = []
    f = open(path, 'r')
    line=f.readline()
    line=f.readline()
    while (line):
        words = line.split(' ')
        if words:
            if (len(words) > 1):
                data.append([words[0],words[1]])
        line=f.readline()
    f.close()
    return data

import fractions

def lcm(a,b): 
    return abs(a * b) / fractions.gcd(a,b) if a and b else 0 

def get_multi(x):
    S = str(x).split(".")
    if (len(S) == 1):
        return 1
    L = 10**len(S[1])
    div = alldividers(L)
    for i in div:
        if(i*x == float(int(i*x))):
           return i
                      
def alldividers(zahl):
    multi = 10.0
    try:
        temp = str(zahl).split(".")
        multi = 10.0**len(temp[1])
    except:
        pass        
    result = []
    for i in range (1, zahl + 1):
        if zahl * multi / i % multi == 0:
            result.append(i)
    return result


def create_structure(path, a, c, a11 ,a12 ,a21 ,a22, l1, l2, bl):
    f = open(path, 'w')
    L1 = get_multi(l1)
    L2 = get_multi(l2)
    L = lcm(L1,L2)
    layers = bl*(l1+l2)
    if (float(int(layers)) == layers):
        layers = int(layers)
    else:
        layers = int(layers)+1
    f.write('a ' + str(a*L*10**(-10)) + '\n')
    f.write('b ' + str(a*10**(-10)) + '\n')
    f.write('c ' + str(c*layers*10**(-10)) + '\n')
    atom = 1
    l = l1
    offset = 0
    for i in range(layers):
        for j in range(L):
            #check that the right atom is printed
            if (i*L+j - offset >= l*L):
                if (l == l1):
                    l = l2
                    atom = 2
                else:
                    l = l1
                    atom = 1
                offset = i*L+j
            if (atom == 1):
                f.write('element ' + a11 + '\n')
            else:
                f.write('element ' + a21 + '\n')
            f.write('x ' + str(float(j)/L) + '\n')
            f.write('y ' + str(0) + '\n')
            f.write('z ' + str(float(i)/layers) + '\n')
            if (atom == 1):
                f.write('element ' + a12 + '\n')
            else:
                f.write('element ' + a22 + '\n')
            f.write('x ' + str((j+0.5)/L) + '\n')
            f.write('y ' + str(0.5) + '\n')
            f.write('z ' + str((i+0.5)/layers) + '\n')
            f.write('element O' + '\n')
            f.write('x ' + str((j+0.5)/L) + '\n')
            f.write('y ' + str(0.5) + '\n')
            f.write('z ' + str(float(i)/layers) + '\n')
            f.write('element O' + '\n')
            f.write('x ' + str((j+0.5)/L) + '\n')
            f.write('y ' + str(0) + '\n')
            f.write('z ' + str((i+0.5)/layers) + '\n')
            f.write('element O' + '\n')
            f.write('x ' + str(float(j)/L) + '\n')
            f.write('y ' + str(0.5) + '\n')
            f.write('z ' + str((i+0.5)/layers) + '\n')
    f.close()
    print "l", L, "layers", layers
    
def showup(l1, l2):
    L1 = get_multi(l1)
    L2 = get_multi(l2)
    L = lcm(L1,L2)
    bl = get_multi((l1+l2))
    layers = float(bl)*(l1+l2)
    
    if (float(int(layers)) == layers):
        layers = int(layers)
    else:
        print "error bilayers wrong"    
    
    atom = 1
    l = l1
    offset = 0
    result = ''
    for i in range(layers):
        for j in range(L):
            #check that the right atom is printed
            if (i*L+j - offset >= l*L):
                if (l == l1):
                    l = l2
                    atom = 2
                else:
                    l = l1
                    atom = 1
                offset = i*L+j
            if (atom == 1):
                result += 'B'
            else:
                result += 'S'
        result += '\n'
    each = result.split('\n')
    temp = []
    for i in range(len(each)-1, -1 ,-1):
        temp.append(each[i])
    for line in temp:
        print line     
```

```python
import math as m
import imp
%pylab inline
import matplotlib.pyplot as plt

xpy = imp.load_source('xpy', '/home/benjaminbein/workespace/git/x-ray_simulator/xpy/src/xpy.py')

l_min = 0.8
l_max = 1.05
l_step = 0.0001

wavelength = 1.5409E-10
direct = 2.0E8
background = 1

path = r'/home/benjaminbein/workespace/git/x-ray_simulator/Perovskites/BTOSTO.str'
create_structure(path, 3.905, 3.988, 'Ba', 'Ti', 'Sr', 'Ti', 2.1, 6.3, 10)

STO = xpy.SimpleCrystalStructure(path = xpy.scriptpath+"Perovskites/STOh.str")
STOs = xpy.SimpleCrystalStructure(path = xpy.scriptpath+"Perovskites/STOh.str")
SRO = xpy.SimpleCrystalStructure(path = xpy.scriptpath+"Perovskites/SROs.str")
BTO = xpy.SimpleCrystalStructure(path = xpy.scriptpath+"Perovskites/PTOs.str")
BTOSTO = xpy.SimpleCrystalStructure(path = xpy.scriptpath+"Perovskites/BTOSTO.str")
substrate = xpy.Film(STO)
electrode = xpy.ThinFilm(SRO,46)
sfilmSTO = xpy.ThinFilm(STOs,6)
sfilmSTO.delta = 0.
sfilmBTO = xpy.ThinFilm(BTO,2)
sfilmBTO.delta = 0.
superlatticesimple = xpy.SuperLattice(sfilmBTO,sfilmSTO,10)
superlattice = xpy.ThinFilm(BTOSTO,1)

q = [0.0,0.0,0.0]
steps_l = (l_max-l_min)/l_step
```

```python
BTO.lattice_parameters[2] = 3.988e-10
STOs.lattice_parameters[2] = 3.988e-10
s101 = []
sub101 = []
q[0] = 2*m.pi/STO.lattice_parameters[0]
for i in range(int(steps_l)):
    l=l_min+i*l_step
    q[2] = l*2*m.pi/STO.lattice_parameters[2]
    sub = substrate.get_reflection(q, wavelength)
    lin = abs(
              superlattice.get_reflection(q, wavelength) + 
              sub * superlattice.get_phase_shift(q, wavelength)
             )
    sub = abs(sub)
    sub *= sub
    sub *= direct
    sub += background
    lin *= lin
    lin *= direct
    lin += background
    s101.append([l, lin])
    sub101.append([l, sub])
#print superlattice.get_c()
#print BTO.lattice_parameters[2]
#print STOs.lattice_parameters[2]
#print superlattice.get_Lambda()
```
