```python
def load_sim(path):
    data = []
    f = open(path, 'r')
    line=f.readline()
    line=f.readline()
    while (line):
        words = line.split(' ')
        if words:
            if (len(words) > 1):
                data.append([words[2],words[3]])
        line=f.readline()
    f.close()
    return data

def load_data(path):
    data = []
    f = open(path, 'r')
    line=f.readline()
    line=f.readline()
    while (line):
        words = line.split(' ')
        if words:
            if (len(words) > 1):
                data.append([words[0],words[1]])
        line=f.readline()
    f.close()
    return data

import fractions
import random

def lcm(a,b): 
    return abs(a * b) / fractions.gcd(a,b) if a and b else 0 

def get_multi(x):
    S = str(x).split(".")
    if (len(S) == 1):
        return 1
    L = 10**len(S[1])
    div = alldividers(L)
    for i in div:
        if(i*x == float(int(i*x))):
           return i
                      
def alldividers(zahl):
    multi = 10.0
    try:
        temp = str(zahl).split(".")
        multi = 10.0**len(temp[1])
    except:
        pass        
    result = []
    for i in range (1, zahl + 1):
        if zahl * multi / i % multi == 0:
            result.append(i)
    return result


def create_structure(path, a, c, a11 ,a12 ,a21 ,a22, l1, l2, bl, zoffset1 = 0.0, zoffset2 = 0.0):
    f = open(path, 'w')
    L1 = get_multi(l1)
    L2 = get_multi(l2)
    L = lcm(L1,L2)
    layers = bl*(l1+l2)
    if (float(int(layers)) == layers):
        layers = int(layers)
    else:
        layers = int(layers)+1
    f.write('a ' + str(a*L*10**(-10)) + '\n')
    f.write('b ' + str(a*10**(-10)) + '\n')
    f.write('c ' + str(c*layers*10**(-10)) + '\n')
    atom = 1
    l = l1
    offset = 0
    zoffset = zoffset1
    for i in range(layers):
        for j in range(L):
            #check that the right atom is printed
            if (i*L+j - offset >= l*L):
                if (atom == 1):
                    l = l2
                    atom = 2
                    zoffset += zoffset2
                else:
                    l = l1
                    atom = 1
                offset = i*L+j
                zoffset += zoffset1
            if (atom == 1):
                f.write('element ' + a11 + '\n')
            else:
                f.write('element ' + a21 + '\n')
            f.write('x ' + str((j+zoffset+random.randrange(-1,1)*30**-11)/L) + '\n')
            f.write('y ' + str(0.0+random.randrange(-1,1)*30**-11) + '\n')
            f.write('z ' + str(float(i+random.randrange(-1,1)*30**-11)/layers) + '\n')
            if (atom == 1):
                f.write('element ' + a12 + '\n')
            else:
                f.write('element ' + a22 + '\n')
            f.write('x ' + str((j+0.5+zoffset+random.randrange(-1,1)*30**-11)/L) + '\n')
            f.write('y ' + str(0.5+random.randrange(-1,1)*30**-11) + '\n')
            f.write('z ' + str((i+0.5)/layers+random.randrange(-1,1)*30**-11) + '\n')
            f.write('element O' + '\n')
            f.write('x ' + str((j+0.5+zoffset+random.randrange(-1,1)*30**-11)/L) + '\n')
            f.write('y ' + str(0.5+random.randrange(-1,1)*30**-11) + '\n')
            f.write('z ' + str(float(i)/layers+random.randrange(-1,1)*30**-11) + '\n')
            f.write('element O' + '\n')
            f.write('x ' + str((j+0.5+zoffset+random.randrange(-1,1)*30**-11)/L) + '\n')
            f.write('y ' + str(0.0+random.randrange(-1,1)*30**-11) + '\n')
            f.write('z ' + str((i+0.5+zoffset+random.randrange(-1,1)*30**-11)/layers) + '\n')
            f.write('element O' + '\n')
            f.write('x ' + str(float(j+random.randrange(-1,1)*30**-11)/L) + '\n')
            f.write('y ' + str(0.5+random.randrange(-1,1)*30**-11) + '\n')
            f.write('z ' + str((i+0.5+zoffset+random.randrange(-1,1)*30**-11)/layers) + '\n')'''
    f.close()
    print "l", L, "layers", layers
    
def showup(l1, l2):
    L1 = get_multi(l1)
    L2 = get_multi(l2)
    L = lcm(L1,L2)
    bl = get_multi((l1+l2))
    layers = float(bl)*(l1+l2)
    
    if (float(int(layers)) == layers):
        layers = int(layers)
    else:
        print "error bilayers wrong"    
    
    atom = 1
    l = l1
    offset = 0
    result = ''
    for i in range(layers):
        for j in range(L):
            #check that the right atom is printed
            if (i*L+j - offset >= l*L):
                if (atom == 1):
                    l = l2
                    atom = 2
                else:
                    l = l1
                    atom = 1
                offset = i*L+j
            if (atom == 1):
                result += 'B'
            else:
                result += 'S'
        result += '\n'
    each = result.split('\n')
    temp = []
    for i in range(len(each)-1, -1 ,-1):
        temp.append(each[i])
    for line in temp:
        print line     
```

```python
import math as m
import imp
%pylab inline
import matplotlib.pyplot as plt

xpy = imp.load_source('xpy', '/home/benjaminbein/workespace/git/x-ray_simulator/xpy/src/xpy.py')

l_min = 0.8
l_max = 1.05
l_step = 0.0001

wavelength = 1.5409E-10
direct = 2.0E8
background = 1

q = [0.0,0.0,0.0]
steps_l = (l_max-l_min)/l_step
```

#101

#Substrate


```python
STO = xpy.SimpleCrystalStructure(path = xpy.scriptpath+"Perovskites/STOh.str")
substrate = xpy.Film(STO)
sub101 = []
q[0] = 2*m.pi/STO.lattice_parameters[0]
q[1] = 0
for i in range(int(steps_l)):
    l=l_min+i*l_step
    q[2] = l*2*m.pi/STO.lattice_parameters[2]
    absq = sum([x**2 for x in q])
    sinomega = 4.0 * m.pi / absq / wavelength
    sub = substrate.get_reflection(q, wavelength, sinomega)
    sub = abs(sub)
    sub *= sub
    sub *= direct
    sub += background
    sub101.append([l, sub])
```

```python
fig = plt.figure()

axes = fig.add_axes([0.0, 0.0, 1.0, 1.0]) # left, bottom, width, height (range 0 to 1)

axes.plot([x[0] for x in subdata], [float(a[1]) for a in subdata], color="green", lw=2, ls='*', marker='+')
#axes.plot([x[0] for x in sub101], [float(a[1]) + b for a, b in zip(sub101, dif) ], color="blue", lw=2, ls='*', marker='+')
axes.plot([x[0] for x in sub101], [float(a[1]*4) for a in sub101], color="red", lw=2, ls='*', marker='+')

axes.set_xlabel('l')
axes.set_ylabel('int[A.U.]')
axes.set_title('title')
axes.set_yscale("log")
```

#Electrode

```python
SRO = xpy.SimpleCrystalStructure(path = xpy.scriptpath+"Perovskites/SROs.str")
electrode = xpy.ThinFilm(SRO,41)
SRO.lattice_parameters[0] = SRO.lattice_parameters[1] = 3.925*10**-10
SRO.lattice_parameters[2] = 4.*10**-10
electrode.delta = 0.0

e101 = []
q[0] = 2*m.pi/STO.lattice_parameters[0]
q[1] = 0
for i in range(int(steps_l)):
    l=l_min+i*l_step
    q[2] = l*2*m.pi/STO.lattice_parameters[2]
    absq = sum([x**2 for x in q])
    sinomega = 0.02*absq/absq #4.0 * m.pi / absq / wavelength
    lin = abs(electrode.get_reflection(q, wavelength, sinomega) +
              substrate.get_reflection(q, wavelength, sinomega) * electrode.get_phase_shift(q, wavelength, sinomega)
             )
    lin *= lin
    lin *= direct
    lin += background;
    e101.append([l, lin])
```
