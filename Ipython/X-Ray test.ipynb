```python
def load_sim(path):
    data = []
    f = open(path, 'r')
    line=f.readline()
    line=f.readline()
    while (line):
        words = line.split(' ')
        if words:
            if (len(words) > 1):
                data.append([words[2],words[3]])
        line=f.readline()
    f.close()
    return data

def load_data(path):
    data = []
    f = open(path, 'r')
    line=f.readline()
    line=f.readline()
    while (line):
        words = line.split(' ')
        if words:
            if (len(words) > 1):
                data.append([words[0],words[1]])
        line=f.readline()
    f.close()
    return data
```

```python
import math as m
import imp
%pylab inline
import matplotlib.pyplot as plt

xpy = imp.load_source('xpy', '/home/benjaminbein/workespace/git/x-ray_simulator/xpy/src/xpy.py')

l_min = 0.8
l_max = 1.05
l_step = 0.0001

wavelength = 1.5409E-10
direct = 2.0E8
background = 1

STO = xpy.SimpleCrystalStructure(path = xpy.scriptpath+"Perovskites/STOh.str")
STOs = xpy.SimpleCrystalStructure(path = xpy.scriptpath+"Perovskites/STOh.str")
SRO = xpy.SimpleCrystalStructure(path = xpy.scriptpath+"Perovskites/SROs.str")
BTO = xpy.SimpleCrystalStructure(path = xpy.scriptpath+"Perovskites/PTOs.str")
substrate = xpy.Film(STO)
electrode = xpy.ThinFilm(SRO,46)
sfilmSTO = xpy.ThinFilm(STOs,6)
sfilmBTO = xpy.ThinFilm(BTO,2)
superlattice = xpy.SuperLattice(sfilmBTO,sfilmSTO,10)

q = [0.0,0.0,0.0]
steps_l = (l_max-l_min)/l_step
```

```python
BTO.lattice_parameters[2] = 3.987e-10
STOs.lattice_parameters[2] = 3.987e-10
s001 = []
q[0] = 0 #2*m.pi/STO.lattice_parameters[0]
for i in range(int(steps_l)):
    l=l_min+i*l_step
    q[2] = l*2*m.pi/STO.lattice_parameters[2]
    lin = abs(
              superlattice.get_reflection(q, wavelength) + 
              substrate.get_reflection(q, wavelength) * superlattice.get_phase_shift(q, wavelength)
             )
    #lin = abs(film.get_reflection(q, wavelength))
    #lin = abs(substrate.get_reflection(q, wavelength))
    lin *= lin
    lin *= direct
    lin += background;
    s001.append([l, lin])
print superlattice.get_c()
print BTO.lattice_parameters[2]
print STOs.lattice_parameters[2]
print superlattice.get_Lambda()
```

```python
BTO.lattice_parameters[2] = 3.973e-10
STOs.lattice_parameters[2] = 3.973e-10
s101 = []
sub101 = []
q[0] = 2*m.pi/STO.lattice_parameters[0]
for i in range(int(steps_l)):
    l=l_min+i*l_step
    q[2] = l*2*m.pi/STO.lattice_parameters[2]
    sub = substrate.get_reflection(q, wavelength)
    lin = abs(
              superlattice.get_reflection(q, wavelength) + 
              sub * superlattice.get_phase_shift(q, wavelength)
             )
    sub = abs(sub)
    sub *= sub
    sub *= direct
    sub += background
    lin *= lin
    lin *= direct
    lin += background
    s101.append([l, lin])
    sub101.append([l, sub])
print superlattice.get_c()
print BTO.lattice_parameters[2]
print STOs.lattice_parameters[2]
print superlattice.get_Lambda()
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)


axes.plot([x[0] for x in s101], [y[1] for y in s101], color="blue", lw=1, ls='*', marker='+')
#axes.plot([x[0] for x in s101], [float(a[1]) + b for a, b in zip(s101, dif) ], color="blue", lw=2, ls='*', marker='+')
#axes.plot([x[0] for x in s001], [y[1] for y in s001], color="green", lw=1, ls='*', marker='+')
#axes.plot([x[0] for x in filmdata], [y[1] for y in filmdata], color="red", lw=1, ls='*', marker='+')
axes.plot([x[0] for x in filmdata], [y[1] for y in filmdata], color="red", lw=1, ls='*', marker='+')

axes.set_xlabel('l')
axes.set_ylabel('int[A.U.]')
axes.set_title('title')
axes.set_yscale("log")
```
